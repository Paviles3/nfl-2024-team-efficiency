```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("NFL_2024_Stats.csv")

# Show the first few rows
df.head()
```

```python
# Check for nulls or weird values
print(df.info())

# Get summary statistics
df.describe()
```

```python
# Top 10 Teams by Points Scored

top_10_pts_scored = df.sort_values("Points Scored", ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(top_10_pts_scored["Team"], top_10_pts_scored["Points Scored"], color="forestgreen")
plt.title("Top 10 Teams by Points Scored (2024)")
plt.xlabel("Team")
plt.ylabel("Points Scored")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
